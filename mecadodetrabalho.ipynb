## Taxa de desocupação - Alagoas/Brasil - 2019/2025

```python
df = pd.read_excel('/content/drive/MyDrive/Boletim de Conjuntura/Dados Boletim/Mercado de trabalho/1T2025_mercadotrabalho.xlsx',
                   sheet_name='tabela4099', skiprows=1)

df.head()
```

```python
df.plot(
    x="Trimestre",
    y=["Brasil", "Alagoas"],
    marker="o",
    figsize=(12, 5),
    color=["tab:blue", "navy"]
)

plt.title("Taxa de desocupação - Alagoas/Brasil - 2019/2025")
plt.xlabel("Trimestre")
plt.xticks(rotation=45)
plt.legend()
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
ax = df.plot(
    x="Trimestre",
    y=["Brasil", "Alagoas"],
    marker="o",
    figsize=(12, 5),
    color=["tab:blue", "navy"]
)

# Adiciona os rótulos(valores) em cada ponto
for line in ax.get_lines():
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x, y + 0.2, f"{y:.1f}", ha='center', fontsize=9)

plt.title("Taxa de desocupação - Alagoas/Brasil - 2019/2025")
plt.xlabel("Trimestre")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## Caged - Setores 1º Tri 2025

```python
df2 = pd.read_excel('/content/drive/MyDrive/Boletim de Conjuntura/Dados Boletim/Mercado de trabalho/1T2025_mercadotrabalho.xlsx',
                   sheet_name='caged')
df2.head()
```

```python
df2.plot(
    x='Setor',
    y='SALDO 1T/25',
    kind='bar',
    figsize=(10, 6),
    color=['cornflowerblue'],
    legend = False
)

plt.title('Saldo de empregos por atividade em Alagoas - 1º Trimeste de 2025')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
ax = df2.plot(
    x='Setor',
    y='SALDO 1T/25',
    kind='bar',
    figsize=(10, 6),
    color='cornflowerblue',
    legend=False
)

plt.title('Saldo de empregos por atividade em Alagoas - 1º Trimestre de 2025')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.axhline(0, color='black', linewidth=1) # Adiciona linha no zero
for container in ax.containers: ax.bar_label(container, fmt='%.0f', label_type='edge', padding=3) # Adiciona rótulos

plt.tight_layout()
plt.show()
```
